```python
import pandas as pd
df = pd.read_csv("superstore.csv")
df.head()
```

```python
east_sales = df[df["Region"] == "East"]["Sales"]
west_sales = df[df["Region"] == "West"]["Sales"]
```

```python
east_sales = df[df["Region"] == "East"]["Sales"]
west_sales = df[df["Region"] == "West"]["Sales"]
```

```python
from scipy import stats
```

```python
t_stat, p_value = stats.ttest_ind(east_sales, west_sales, equal_var=False)

t_stat, p_value
```

```python
alpha = 0.5

if p_value < alpha:
    print("Reject the Null Hypothesis")
else:
    print("Fail to Reject the Null Hypothesis")
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
plt.figure(figsize=(8,5))
sns.histplot(east_sales, color="blue" , label="East" , kde= True)
sns.histplot(west_sales, color="Orange" , label= "West" , kde= True)
plt.legend()
plt.title("Sales Distribution: East vs West")
plt.show()
```

```python
plt.figure(figsize=(6,5))
sns.boxplot(x="Region", y="Sales", data=df[df["Region"].isin(["East","West"])])
plt.title("Sales Comparison between East and West")
plt.show()
```

```python
east = df[df["Region"]=="East"]["Sales"]
west = df[df["Region"]=="West"]["Sales"]
south = df[df["Region"]=="South"]["Sales"]
central = df[df["Region"]=="Central"]["Sales"]

f_stat, p_val = stats.f_oneway(east, west, south, central)
f_stat, p_val
```
